```python
import pandas as pd 
```

```python
data = pd.read_csv('./Iris.csv')
data
```

```python
def label_encoder(data):
    for col in data.columns:
        if data[col].dtype == 'object':
            data[col] = pd.factorize(data[col])[0]
    return data
```

```python
label_encoder(data).head(100)


```
